```python
def percent_missing(df):
    percents_nan = 100*df.isnull().sum()/len(df)
    percents_nan = percents_nan[percents_nan>0].sort_values()
    
    return percents_nan
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
plt.figure(figsize=(8,5),dpi=200)
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.xticks(rotation=90); # gira os nomes do gráfico em 90º
plt.ylim(0,1)
```
